<h3>Since database smells can only occur on files that have database acess. All classes with dbAccess less than one needs to be filtered out.</h3>

```python
#filtering files with dbAccess greater than one
df=df[df['dbAccess']>0]
media=df[df['category']=='Multimedia']
business= df[df['category']=='Business']
utility=df[df['category']=='Utility']
library=df[df['category']=='Library']
```

```python
df.groupby('category')['category'].value_counts()
```

<h1>Apriori Algorithm on the overall dataset <h1>

```python
df.columns
```

```python
#now lets filterout the columns that are not needed for the association rule mining
fil_df =  df.drop(['category', '_date', 'version', 'projectGroup', 'projectName',
       'file_id', 'class_path', 'dbAccess','in_bug_fix', 'in_bug_induce',], axis=1)
fil_df.head()
```

<h3>For association rule mining, the values in all columns must be either zero(present) or one(absent). For some smells like longMethod a class may contain more than one value. So any value greater than zero has to be converted to 1(present)</h3>

```python
fil_df[fil_df > 0]=1
```

```python
fil_df['IMPLICIT_COLUMNS'].value_counts()
```

```python
fil_df['FEAR_OF_THE_UNKNOWN'].value_counts()
```

```python
fil_df['AMBIGUOUS_GROUPS'].value_counts()
```

```python
fil_df['RANDOM_SELECTION'].value_counts()
```

```python
fil_df['LongMethod'].value_counts() #sample traditional smell
```

```python
frequent_itemsets=apriori(fil_df,min_support=0.01, max_len=2, use_colnames=True)
frequent_itemsets.head(100)
```

```python
#geneate the rules
rules=association_rules(frequent_itemsets, metric="lift",min_threshold=1)
```

```python
rules=rules.sort_values(by=['lift'], ascending=False)
rules.to_csv("Rules_combined.csv", encoding='utf-8', index=False)
rules
```
